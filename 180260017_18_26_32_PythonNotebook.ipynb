## Motivation and Introduction

### What is the paper about?

The paper our group will be discussing is a June 1999 paper by Ryan Bennink, Young-Kwon Yoon and Robert W. Boyd of the University of Rochester, and J.E. Sipe of the University of Toronto. Titled 'Accessing the optical nonlinearity of metals with metal-dielectric photonic bandgap structures', the paper describes a novel structure, that they numerically show to have a very high nonlinear refractive index, whilst being reasonably transmissive.$^1$

The basic idea behind the paper is that metals like Copper, while having very high nonlinear susceptibilities (~$10^{6}$ times that of usual dielectrics) are (or rather, were) of no use to us because of them being nearly opaque. To access these nonlinear properties of metals, while having a structure be able to transmit reasonably well, was the goal of this paper.<br>
The structure that the authors propose is a multilayer metal-dielectric structure, with the metal's layer thickness being relatively smaller, and the dielectric comprising the bulk of the material (the actual structure proposed shall be shown later in the notebook). <br>
The authors numerically show that this proposed structure has a high 3<sup>rd</sup> order nonlinear susceptibility while maintaining moderate transmission.

### Great! But, why is this important?

A good question to have at this point, is why exactly was there a need for this paper? Why do we care about high nonlinear susceptibilities at all? Well, throughout the course, we've looked at places where we use nonlinear properties of materials in places such as second-harmonic generation (not relevant to this paper, since we only look at third order effects), third-harmonic generation, four-wave mixing, intensity dependence of refractive index (main metric we'll be looking at in the paper to gauge our structure's properties), etc.

With that in mind, having materials that give you higher non-linear susceptibilities with reasonable transmission is a pretty big breakthrough. The structure proposed in the paper allows to access the high nonlinear susceptibilities of metals, and allow us to create better materials for usage in cheaper lasers, say. This allows for experiments to be more affordable, and paves the way for mass-production of cheaper, better nonlinear materials for the future.

### Sounds good. How exactly do the authors do this?

The paper mentions alternate studies that have been done into the field of creating metal-dielectric composites that enhance nonlinear polarization and keep absorption at an acceptable level. Examples are made of Maxwell Garnett and Bruggeman geometries, in which local-field effects are utilized to achieve this. The paper gives a novel approach, utilizing a layered metal-dielectric geometry, which makes use of photonic band resonance to maximise transmission and increase non-linear phase shift. 

The authors make a crucial distinction here. The attenuation constant of a material is proportional to the imaginary part of the refractive index, while the absorption constant is proportional to the imaginary part of the dielectric constant. In metals, the imaginary part of the refractive index, is typically larger than the imaginary part of the dielectric constant. This leads to the implication that the large attenuation of light in metals is more due to reradiation in the backward direction, than due to absorption.

The authors then make the claim that if this light could be redirected in the forward direction through methods such as Bragg reflection, this could greatly increase the transmission of the sample. The actual method used to achieve this, is of using photonic bandgap (PBG) structures to increase the transmittance of a given thickness of metal. 

This method was demonstrated and conceived of, by Mark J. Bloemer and his co-workers. In a 1998 paper titled 'Transparent, metallo-dielectric, one-dimensional, photonic band-gap structures' by M Scalora, MJ Bloemer, AS Pethel, JP Dowling, CM Bowden and AS Manka, they investigated numerically, the properties of metallo-dielectric, one-dimensional, PBG structures. They showed how these structures can be made transparent for a tunable range of frequencies, remaining transparent even when the total metal thickness is increased to hundreds of skin depths in net thickness.$^2$

In a paper dated a month later, titled 'Transmissive properties of $Ag/MgF_2$ photonic band gaps' by MJ Bloemer and M Scalora, they presented their results of transmittance measurements on periodic layers of $Ag/MgF_2$ at optical and microwave frequencies. These papers form the basis for making the structures mentioned in our main paper, with their increased transmission and high nonlinear susceptibilities.$^{11}$

### Cool! So what is recreated in this notebook?

We recreate here exactly what's mentioned in the paper. We numerically solve for Maxwell's equations with appropriate boundary conditions for the metal-dielectric PBG structure, and show how the increased transmission that these materials offer leads to intensity-dependent changes in the complex amplitude of a transmitted beam, that are much larger than in a bulk sample of the same thickness of metal. In short, we show why these structures have higher nonlinear susceptibilities and second order refractive indices, whilst showing how they offer reasonable transmission through them.

## Problem and methodology
### Geometry and material properties

First and foremost, the geometry. The PBG structure consists of alternating layers of $SiO_2$ (100nm) and $Cu$ (16nm) with the first layer being $SiO_2$. The whole structure starts at $z=0$ and ends at $z=L$. The assumptions made about the materials are the following:
- both $SiO_2$ and $Cu$ are non-magnetic
- no second order susceptibilities
- copper has the dominant $\chi^{(3)}$

### Experimental Setup and relevant equations
We assume the PBG crystal is aligned with its normal along the $z$-direction, and is infinite in the $x$ and $y$ directions. An $x$-polarized electromagnetic wave is incident at $z = 0$, the beginning of the crystal:

$$\mathbf{E}=E_{i}e^{i(kz -\omega t)}\;\hat{x} \ , \ \mathbf{H}=H_{i}e^{i(kz -\omega t)}\;\hat{y} \ .$$ 

Maxwell's equations (in SI units) describing E-M waves inside the crystal are given by:

$$\nabla\times \mathbf{E}=-\mu_{0}\frac{\partial\mathbf{H}}{\partial t} \ ,$$
$$\nabla\times \mathbf{H}=J+\frac{\partial\mathbf{D}}{\partial t} \ ,$$
$$\nabla\cdot \mathbf{H}=0 \ ,$$
$$\nabla\cdot \mathbf{D}=0 \ .$$

Note that there is no free charge density. Due to the continuous translation invariance of the system in the $x$ and $y$ directions, the electric and magnetic fields retain their polarizations thoughout the entire space. Also the nonlinear contribution is tiny compared to the linear terms, so the time dependance can be safely approximated as $e^{-i\omega t}$. This ensures that all time derivatives in Maxwell's equations are simply replaced by the factor $-i\omega$. Hence the complex electric and magnetic fields in the crystal take the form:

$$\mathbf{E}=E(z)e^{-i\omega t}\;\hat{x} \ , \ \mathbf{H}=H(z)e^{-i\omega t}\;\hat{y} \ .$$

The electric displacement field $\mathbf{D}$ in either regions of the crystal is given by:

$$\textbf{D}=\epsilon_0\epsilon_r\mathbf{E}+\epsilon_0\chi^{(3)}_{1111}\mathbf{E}^3$$

We use the above two approximations and substitute them in Maxwell's equations to obtain the coupled ODEs for the fields inside the crystal (in SI units) :

$$dE/dz = i\mu_0\omega H(z) \ ,$$
$$dH/dz = -(i\epsilon_0\omega-\sigma)E(z)+3i\epsilon_0\omega\chi^{(3)}_{1111}E(z)^3 \ ,$$

We make the final substitution: $\epsilon_0 + i\sigma / \omega = \epsilon_0 n_{Cu}^2$, to obtain the ODEs in the paper in SI units. 
Finally, we supply the appropriate boundary conditions on either side of the PBG.

$$E_i = \frac{1}{2} [E(0) + \mu_0 cH(0)]~,$$
$$E_r = \frac{1}{2} [E(0) - \mu_0 cH(0)~,$$
$$E_t = \frac{1}{2} [E(L) + \mu_0 cH(L)]~,$$
$$0 = \frac{1}{2} [E(L) - \mu_0 cH(L)]~.$$

We are now ready to solve for E and H. Due to the non-linearity in the ODEs, the solutions are not analytically obtainable. So we solve these numerically

## Setup

We begin by importing necessary constants. The paper uses primarily Gaussian (CGS) units, but we made the choice to instead prefer SI through the demonstration.

```python
from numpy import pi, sqrt
from scipy.constants import speed_of_light as c_vac, mu_0, epsilon_0
```

With universal constants in place, we set our parameters.

```python
# parameters

    ## refractive indices
n_Cu =        0.216 + 3.64j
n_SiO2 =      1.456
n_air =       1

    ## 3rd order susceptibilities
chi_Cu =      1.668E-18
chi_SiO2 =    0
chi_air =     0
    
    ## light properties
wavelength = 650e-9
ang_freq = (2*pi * c_vac) / wavelength

    ## material configurations
len_SiO2 =    100e-9
len_Cu   =    80e-9
len_air =     1.5*len_SiO2

# configuration vectors, for use with solver
params_Cu =   [n_Cu, chi_Cu]
params_SiO2 = [n_SiO2, chi_SiO2]
params_air =  [n_air, chi_air]
```

And the following conditions for the differential equations, to be numerically solved

```python
# ode params

    ## initial values, corresponding to a purely transmitted wave
E_t =   1.0
H_t =   -1.0 / (mu_0 * c_vac)
y0 =   [E_t, H_t]   # actual system variable vector

    ## solver parameters; limits and step-size
zStop = len_air
zInc =  0.5e-10
z0 =    0

    ## containers for solver results
zvals = []
Vals =  []
```

```python
# System of Equations

    ## first derivatives
def f(z, y, params):
    E, H = y
    n_linear, chi = params

    derivs = [1j * ang_freq * mu_0 * H , 
             (1j * ang_freq * epsilon_0 * n_linear**2) * E + 1j * chi * (E**3) ]

    return derivs


    ## jacobian 
def jac(z, y, params):
    E, H = y
    n_linear, chi = params

    jacobian = [
                    [0, 1j*ang_freq*mu_0], 
                    [(1j * ang_freq * epsilon_0 * (n_linear ** 2)) + 3j * chi * (E**2), 0]
                ]

    return jacobian
```

Finally, setting up the numerical solver with our parameters

For numerically solving the ODEs, we make use of `` scipy.integrate.ode `` . The solver numerically solves the ODEs, using a user-inputted Jacobian, and a method that we can specify.

The two methods that can be used for solving a system of ODEs are `` 'adams' or 'bdf' `` . The `` 'adams' `` method is primarily used to solve non-stiff ODEs, with `` 'bdf' `` used in the case of stiff ones.

Our solver uses the `` 'bdf' `` method, since we didn't know at the outset whether the system of ODEs is stiff/non-stiff. However, changing the method doesn't change the results within our resolution, so chose to use it for all of the equations.

Another key point to note is that we integrate backwards. Our boundary conditions are easier to implement in the form of there being no incident wave from the right, so beginning from the right side makes it easier for us to use appropriate boundary conditions.

```python
import scipy.integrate as integ

# actual solving - integrating backwards
#   solve in air following the material
#   solve in material
#   solve in air preceding the material

def solve(ode_params, mat_params, len_params):
    y0, z0, zStop = ode_params
    params_air, params_Cu = mat_params
    len_air, len_Cu = len_params

    solver = integ.ode(f, jac).set_integrator('zvode', method='bdf')
    solver.set_initial_value(y0, z0).set_f_params(params_air).set_jac_params(params_air)

    while solver.successful() and solver.t < zStop:
        solver.integrate(solver.t + zInc)
        zvals.append(solver.t)
        Vals.append(solver.y)

    ## set initial conditions for next integration
    y0 = Vals[-1] 
    z0 = zStop      
    zStop += len_Cu

    ## integrating within the material
    solver = integ.ode(f, jac).set_integrator('zvode', method='bdf')
    solver.set_initial_value(y0, z0).set_f_params(params_Cu).set_jac_params(params_Cu)

    while solver.successful() and solver.t < zStop:
        solver.integrate(solver.t + zInc)
        zvals.append(solver.t)
        Vals.append(solver.y)
    ## deriving incident EM wave from backward integration of transmitted wave
    y_i = Vals[-1]
    E_i = 0.5 * (y_i[0] - (y_i[1] * mu_0 * c_vac))
    H_i = -1 * E_i / (mu_0 * c_vac)
    y0 = [E_i, H_i]

    ### set initial conditions for next integration
    y0 = Vals[-1]
    z0 = zStop
    zStop += len_air

    ## deriving incident EM wave from backward integration of transmitted wave
    y_i = Vals[-1]
    E_i = 0.5 * (y_i[0] - (y_i[1] * mu_0 * c_vac))
    H_i = -1 * E_i / (mu_0 * c_vac)
    y0 = [E_i, H_i]

    solver = integ.ode(f, jac).set_integrator('zvode', method='bdf')
    solver.set_initial_value(y0, z0).set_f_params(params_air).set_jac_params(params_air)

    while solver.successful() and solver.t < zStop:
        solver.integrate(solver.t + zInc)
        zvals.append(solver.t)
        Vals.append(solver.y)

solve([y0, z0, zStop], [params_air, params_Cu], [len_air, len_Cu])
```

We obtain the final fields, in terms of their amplitudes of oscillation at each point.

```python
# Final global E & H fields, reduced to magnitudes
EVals = [abs(element[0]) for element in Vals]
HVals = [abs(element[1]) for element in Vals]
```

However, due to integrating backwards, these values are in the wrong order as well, so reversing the lists

```python
EVals.reverse()
HVals.reverse()
```

The plot is set up to show the Electric Field Strength as a function of distance. The different regions are clearly demarcated, and the legend makes everything clear.

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

def plot_solution():

    # Plot details
    fig = plt.figure(1, figsize = (8,8))
    ElectricFieldLine, = plt.plot(zvals, EVals, color='orangered',label = 'Electric Field Strength')
    plt.xlabel('Distance')
    plt.ylabel('Electric Field Strength')

    # Legend and Added Plot Details

    x_vertical = len_air
    plt.axvline(x = x_vertical, color='k')
    plt.axvspan(x_vertical, x_vertical + len_Cu, color='dimgray', alpha=0.5, lw=0)
    x_vertical = x_vertical + len_Cu
    plt.axvline(x = x_vertical, color='k')

    darkgray_patch = mpatches.Patch(color='dimgray', label='Copper region')
    first_legend = plt.legend(handles=[ElectricFieldLine], loc='lower left')
    ax = plt.gca().add_artist(first_legend)

    plt.legend(handles = [darkgray_patch], bbox_to_anchor = (0., 1.02, .3, .102), loc= 'lower left')

    axes = plt.gca()
    #axes.set_ylim([0,300])

    # finally, printing
    plt.show()

plot_solution()
```

For our first graph, we need to calculate the transmission coefficient.

This is done by calculating $E_i$ using the boundary conditions for the differential equations.
Since we are integrating bbackwards, the required values $E(0)$ and $H(0)$ will be stored in the end of the list `Vals`.

Using these values, we calculate $E_i$ and subsequently calculate $|t|^2$.

Finally we plot the obtained values of $|t|^2$ vs `len_SiO2`

```python
import numpy as np

len_SiO2 = np.arange(50e-9, 250e-9, 0.5e-9)

# Initial values, corresponding to a purely transmitted wave
E_t =  1.0
H_t = -1.0 / (mu_0 * c_vac)
y0 =   [E_t, H_t]
t = []
zInc =  0.1*pow(10,-10)
zStop = len_air
z0 =    0
```

```python
# Finding trans as a function of len_SiO2
def solve_for_len(len):
    Vals = []

    # ODE Solver 
    ## integrating in air region succeding the crystal
    z0 = 0    
    zStop = len_Cu
    y0 =   [E_t, H_t]
    
    ## integrating within the crystal
    for i in range(5):
        ### integrating in copper
        r = integ.ode(f, jac).set_integrator('zvode', method='bdf')
        r.set_initial_value(y0, z0).set_f_params(params_Cu).set_jac_params(params_Cu)
        while r.successful() and r.t < zStop:
            r.integrate(r.t+zInc)
            Vals.append(r.y)
        ### set initial conditions for next integration
        y0 = Vals[-1]
        z0 = zStop
        zStop += len

        ### integrating in silica
        r = integ.ode(f, jac).set_integrator('zvode', method='bdf')
        r.set_initial_value(y0, z0).set_f_params(params_SiO2).set_jac_params(params_SiO2)
        while r.successful() and r.t < zStop:
            r.integrate(r.t+zInc)
            Vals.append(r.y)
        ### set initial conditions for next integration
        y0 = Vals[-1]
        z0 = zStop
        zStop += len_Cu

    ## deriving incident EM wave from 
    ## backward integration of transmitted wave
    ## and deriving transmission coefficient
    y_i = Vals[-1]
    E_i = 0.5 * (y_i[0] - (y_i[1] * mu_0 * c_vac)) 
    trans = (abs(E_t/E_i))**2
    return trans
```

All we're doing in this section is iterating over the various selected values of length of $SiO_2$ in the structure, and compute the transmission ratio for each of them.

However, this proves to be too slow, but since the computations for individual silica lengths are independent, they end up being easily parallelizable. Scheduling the jobs, the following will separate the memory of each execution and run them as separate processes.

```python
from joblib import Parallel, delayed
import multiprocessing

num_cores = multiprocessing.cpu_count()

t = Parallel(n_jobs=num_cores)(delayed(solve_for_len)(len) for len in len_SiO2)
```

```python
# Plot details
fig = plt.figure(1, figsize = (8,8))
plt.plot(len_SiO2,t)
#plt.plot(zvals, HVals)

## plot details
plt.xlabel("len_SiO2")
plt.ylabel("|E_t/E_i|^2")
plt.grid()
plt.show()
```

Our denominator |E_i|^2 is obtained in a similar method to that used for the previous graph.
To calculate the average square of the modulus of the amplitude of the electric field in Cu i.e, <|E_Cu|^2>, first we add all the calculated values of |E_Cu|^2 (obtained from the ode solver).

We now move on to computing something that's of immense interest to us - the ratio of the average electric field intensity within copper layers inside the metal-dielectric structure, to the incident electric field intensity.

Our denominator $|E_i|^2$ is obtained in a similar method to that used for the previous graph.
To calculate the average square of the modulus of the amplitude of the electric field in Cu i.e, $\langle |E_{Cu}|^2 \rangle$, first we add all the calculated values of $|E_{Cu}|^2$ (obtained from the ode solver).
Secondly, we keep track of the number of $|E_{Cu}|^2$ that were added to the aforementioned sum.
Then $\langle |E_{Cu}|^2 \rangle$ is calculated by simply dividing the sum of $|E_Cu|^2$ by the number of $|E_{Cu}|^2$ that were added to make that sum.
Finnaly we plot $\frac{\langle |E_{Cu}|^2 \rangle}{|E_i|^2}$ vs `len_SiO2`

The reason this is essential is because all non-linearities in the PBG structure will arise as a direct consequence of the third-order non-linear susceptibility of the copper layers inside the structure. As a result, the non-linear phase shift (something that's our figure of merit for gauging the non-linearities of the PBG structure) will be directly related to the electric field intensities within the copper layers.

We look at the terminologies used and the derivations later when we compute the graph for the figure of merit.

```python

# Initial values, corresponding to a purely transmitted wave
E_t =  1.0
H_t = -1.0 / (mu_0 * c_vac)
y0 =   [E_t, H_t]
zInc =  1e-11
zStop = len_air
z0 =    0
e_metal_ms = []
# Finding t as a function of len_SiO2


def e_for_len(len):
    Vals = []
    # ODE Solver 
    ## integrating in air region succeding the crystal
    z0 = 0    
    zStop = len_Cu
    y0 =   [E_t, H_t]
    e_metal_s = 0
    count = 0
    ## integrating within the crystal
    for i in range(5):
        ### integrating in copper
        solver = integ.ode(f, jac).set_integrator('zvode', method='bdf')
        solver.set_initial_value(y0, z0).set_f_params(params_Cu).set_jac_params(params_Cu)
        while solver.successful() and solver.t < zStop:
            solver.integrate(solver.t+zInc)
            Vals.append(solver.y)
            e_metal_s += abs(solver.y[0]) ** 2
            count += 1
        ### set initial conditions for next integration
        y0 = Vals[-1]
        z0 = zStop
        zStop += len

        ### integrating in silica
        solver = integ.ode(f, jac).set_integrator('zvode', method='bdf')
        solver.set_initial_value(y0, z0).set_f_params(params_SiO2).set_jac_params(params_SiO2)
        while solver.successful() and solver.t < zStop:
            solver.integrate(solver.t+zInc)
            Vals.append(solver.y)
            
        ### set initial conditions for next integration
        y0 = Vals[-1]
        z0 = zStop
        zStop += len_Cu

    ## deriving incident EM wave from 
    ## backward integration of transmitted wave
    ## and deriving transmission coefficient
    y_i = Vals[-1]
    E_i = 0.5 * (y_i[0] - (y_i[1] * mu_0 * c_vac)) 
    return (e_metal_s/((count)*abs(E_i**2)))

e_metal_ms = Parallel(n_jobs=num_cores)(delayed(e_for_len)(len) for len in len_SiO2)
```

We now move on to the figure of merit- the absolute value of the ratio $\Delta\phi_{PBG} / \Delta\phi_{bulk}$. This involves computation of what exactly the $\Delta\phi$ here is.

The authors define $t = exp(i\phi)$. The phi here is the phase difference between the incident wave and the transmitted wave. However, this $\phi$, which would normally only be due to the first order refractive indices of a material, has a non-linear contribution, because of the intensity-dependent refractive index of copper.

Thus, the authors define $\Delta\phi$ as the intensity-dependent change in complex phase of the transmitted beam. There are a few definitions to unpack here, and a few derivations to make.
We know what the intensity-dependent change in refractive index of a material is.

\begin{equation}
\Delta n = \frac{3}{8n_0}\chi^{(3)}|E|^2\; ,\quad
I = \frac{1}{2}c*n_0*\epsilon_0*|E|^2\; ,\quad
\therefore \Delta n = \frac{3\chi^{(3)}}{4n_0^2\epsilon_0c}I = n_2I
\end{equation}

We have the electric field intensities with us, so we'll use the first formula. But it does help to know the formula in terms of the intensity as well.

Now, 

$$n(I) = n_0 + n_2I$$

By the use of this equation, the phase shift of a plane wave that passes through a third-order nonlinear Kerr medium
with a geometrical length d and an optical length $L_{opt} = nd$, can be calculated as - 

\begin{equation}
\Phi = nL_{opt} = kn_0d + kn_2dI\; ,\quad
\Delta\phi = kn_2dI
\end{equation}

Here, $k = 2\pi / \lambda$. The $\Delta\phi$ thus becomes the intensity dependent change in the phase of the transmitted beam. This comes about as a direct consequence of the non-linearities in the system. In terms of electric field intensity (what we have) - 

$$\Delta\phi = \frac{2\pi}{\lambda}\frac{3d}{8n_0}\chi^{(3)}|E|^2$$

Our electric field intensity isn't constant inside the non-linear medium. And thus, this direct product becomes an integral over $dz$ (in our case)

$$\Delta\phi = \frac{2\pi}{\lambda}\int_{0}^{L}\frac{3}{8n_0}\chi^{(3)}|E|^2 dz$$

We note that $\chi^{(3)}$ is 0 for $SiO_2$ and thus it basically becomes a matter of integrating over the copper layers. This takes a similar form to the averaging of the electric field strength over the copper layers, apart from a scale factor, and thus the graphs are very similar.


```python

# Initial values, corresponding to a purely transmitted wave
E_t =  1.0
H_t = -1.0 / (mu_0 * c_vac)
dp = []
zInc =  1e-11

z0 = 0
zStop = 80e-9
y0 =   [E_t, H_t]
params = params_Cu
delta_phi_Cu = 0
Vals = []
to_integrate = []
zvals = []

k = (2 * pi)/wavelength
n_2_Cu = (3*chi_Cu)/(8*(n_Cu))

solver = integ.ode(f, jac).set_integrator('zvode', method='adams')
solver.set_initial_value(y0, z0).set_f_params(params).set_jac_params(params)

while solver.successful() and solver.t < zStop:
    solver.integrate(solver.t+zInc)
    to_integrate.append( k*n_2_Cu*((abs(solver.y[0]))**2) )
    zvals.append(solver.t)
    Vals.append(solver.y)

y_i = Vals[-1]
E_i = 0.5 * (y_i[0] - (y_i[1] * mu_0 * c_vac) )

delta_phi_Cu = np.trapz(to_integrate, zvals) 
delta_phi_Cu = delta_phi_Cu/((abs(E_i))**2)

def dp_for_len(len):
    Vals = []
    zvals = []
    to_integrate = []

    # ODE Solver 
    ## integrating in air region succeding the crystal
    z0 = 0    
    zStop = len_Cu
    y0 =   [E_t, H_t]

    delta_phi_pbg = 0
    
    ## integrating within the crystal
    for _ in range(5):
        ### integrating in copper
        solver = integ.ode(f, jac).set_integrator('zvode', method='bdf')
        solver.set_initial_value(y0, z0).set_f_params(params_Cu).set_jac_params(params_Cu)
        while solver.successful() and solver.t < zStop:
            solver.integrate(solver.t+zInc)
            Vals.append(solver.y)
            to_integrate.append( k*n_2_Cu*((abs(solver.y[0]))**2) )
            zvals.append(solver.t)
        ### set initial conditions for next integration
        y0 = Vals[-1]
        z0 = zStop
        zStop += len

        ### integrating in silica
        solver = integ.ode(f, jac).set_integrator('zvode', method='bdf')
        solver.set_initial_value(y0, z0).set_f_params(params_SiO2).set_jac_params(params_SiO2)
        while solver.successful() and solver.t < zStop:
            solver.integrate(solver.t+zInc)
            Vals.append(solver.y)
            to_integrate.append(0)
            zvals.append(solver.t)
        ### set initial conditions for next integration
        y0 = Vals[-1]
        z0 = zStop
        zStop += len_Cu
    
    y_i = Vals[-1]
    E_i = 0.5 * (y_i[0] - (y_i[1] * mu_0 * c_vac))

    delta_phi_pbg = np.trapz(to_integrate, zvals)
    delta_phi_pbg = delta_phi_pbg/(abs(E_i))**2
    
    return abs(delta_phi_pbg/delta_phi_Cu)

num_cores = multiprocessing.cpu_count()

dp = Parallel(n_jobs=num_cores)(delayed(dp_for_len)(len) for len in len_SiO2)
```

The fact that we can obtain a figure of merit this high, conclusively shows that the PBG structure detailed in this paper gives a much higher intensity-dependent change in complex amplitude of a transmitted wave, than a bulk sample of copper of same thickness. The PBG structure also has much higher transmission.

## Related work

### It's been a while, have we gotten anything out of it?

Nanolayered composites have been recently proposed to serve as negative index systems, super- and hyperlenses, optical switches, and other nanophotonic structures.$^3$

#### Hyperlenses

For a long time, it was thought that imaging was going to be limited by the diffraction limit. Then, as the science of meta-materials matured, it was discovered that these could be used to make lenses capable of enhancing the exponentially decaying evanescent waves. These were dubbed superlenses. However, these superlenses remained short-sighted, in a sense, capable of near field imaging but not of projecting it onto a far field. Thus, the search continued for suitable *hyper* lenses, which wuould be capable of the same. These heavily utilise metamaterials and their non-linear optical properties, with several proposed designs relying on nanolayer composites as discussed here. In particular, many hyperlens discussions seem to revolve around those made from Silver/Alumina, compared to the Copper/Silica discussion in this paper.$^{3, 5}$

Wang et al (2007) fabricate a hyperlens using a nanolayer composite similar to the setup here, using graphene and boron nitride, utilising a cylindrical geometry for lensing. In particular they utilise an extension of the current system, a two-dimensional model, with the radial and tangential parts behaving differently. It is presented as a possible path for realtime sub-diffraction limit imaging.$^4$

These hyperlens designs would shine in a biological setting, where objects of keen interest, such as viruses and DNA, elude capture even by the most powerful microscopes utilizing traditional imaging. Non-invasive imaging of these opens up a plethora of options for studying them for better understanding and treatment, and a ringer in today's time, vaccine development.$^3$


#### Optical Switching

Another important use of non-linear phase shifting materials comes with optical switching. Interferometer setups with a known phase shift device instead of a detection sample form controllable interference based switches. In particular, the Mach Zehnder Interferometer setup has been widely utilized for switches$^6$

It was long believed that the relaxation time of electrons in a material would limit the switching time that could be reached with it. But Ueno et al (2002) show that placing these phase shifting devices in certain setups, such as the symmetric Mach Zehnder (SMZ) apparatus, leads to the interference masking the relaxation time in a way, with control times being limited by the width of the control pulse rather than the relaxation of the material. However, in such setups, the power required for switch control in these setups can be rather high, Ueno et al citing regular figures of 200W for $AlGaAs$ based switches, which is manageable for a laboratory, but it is a far cry from production and wide spread application.$^6$

With materials as discussed, it could be possible, due to both the high transmission and the phase shift, to reduce the amount of material required, hence making smaller switches, and make them a lot more efficient.

As mentioned, Wang et al (2007)$^4$ talk about a hyperlens made of a similar nanolayer composite, theirs utilising a cylidrical geometry for the sake of lensing. More interesting to our discussion is their material of choice -- the graphene-BN composite. Several studies have been done on tunable transmittance of graphene, such as by Choi et al (2017).$^8$ A combination of these could lead to interesting highly tunable non linear phase shifting devices, and in succession, optical switches. 

Early this year, in a paper published in Nature, Japanese scientists Ono et al demonstrate what they claim to be the fastest all-optical switch by orders of magnitude, using graphene based waveguides. While their results rely on a novel geometry, it goes to show graphene composites have great potential in this application.$^9$
